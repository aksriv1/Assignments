```python
import zipfile

from PIL import Image
from PIL import ImageDraw
import pytesseract
import cv2 as cv
import numpy as np
import math
from kraken import pageseg

# loading the face detection classifier
face_cascade = cv.CascadeClassifier('readonly/haarcascade_frontalface_default.xml')

# the rest is up to you!
```

```python
#Define a function
def zip_images_extraction(name):
    zip_name = "readonly/" + name
    output = []
    # indexing out the information
    with zipfile.ZipFile(zip_name) as myzip:
        zip_infos = myzip.infolist()
        for element in zip_infos:
            name = element.filename
            img = Image.open(myzip.open(name))
            img_strings = pytesseract.image_to_string(img.convert("L"))
            # test if Christopher or Mark are in the text/strings
            if ("Christopher" in img_strings) or ("Mark" in img_strings):
                my_dic = {"name":name, "img":img, "text":img_strings}
                output.append(my_dic)
    return output
```

```python
#extract all information related to small and big images
small_imgs = zip_images_extraction("small_img.zip")
```

```python
big_imgs = zip_images_extraction("images.zip")
```

```python
#define a function to extract a list of faces and 2)create a contact sheet for these faces
def extract_faces(img, scale_factor):
    """
    gray is in array form
    """
    #extract rectangle of faces
    gray = np.array(img.convert("L"))
    faces = face_cascade.detectMultiScale(gray, scale_factor)
    #if no faces are detected
    if len(faces) == 0:
        return None
    #extract faces in a list
    faces_imgs = []
    for x, y, w, h in faces:
        faces_imgs.append(img.crop((x, y, x+w, y+h)))
        
        
    #compute nrows and ncols
    ncols = 5
    nrows = math.ceil(len(faces)/ ncols)
    
    # contact sheet
    contact_sheet = Image.new(img.mode, (500, 100*nrows))
    x, y = (0, 0)
    
    for face in faces_imgs:
        face.thumbnail((100, 100))
        contact_sheet.paste(face, (x,y))
        
        if x+100 == contact_sheet.width:
            x = 0
            y = y + 100
        else:
            x = x + 100
    return contact_sheet
```

```python
#define search function
def value_search(value, zip_name, scale_factor):
    if zip_name == "small_img.zip":
        ref_imgs = small_imgs
    else:
        ref_imgs = big_imgs
    for element in ref_imgs:
        if value in element["text"]:
            print("Results found in file {}".format(element["name"]))
            img = element["img"]
            contact_sheet = extract_faces(img, scale_factor)
            if contact_sheet is not None:
                display(contact_sheet)
            else:
                print("There were no faces in the file")
```

```python
# search for Christopher
value = "Christopher"
zip_name = "small_img.zip"
value_search(value, zip_name, scale_factor = 1.5)
```

```python
# search for Mark
value = "Mark"
zip_name = "images.zip"
value_search(value, zip_name, scale_factor = 1.5)
```
